Practice Filtering with Indian Food Dishes


```python
import pandas as pd
data = {
    'Dish': ['Chole Bhature', 'Masala Dosa', 'Paneer Butter Masala', 'Rajma Chawal', 'Baingan Bharta', 'Pav Bhaji'],
    'Type': ['North', 'South', 'North', 'North', 'North', 'West'],
    'Spice_Level': ['High', 'Medium', 'Medium', 'High', 'Medium', 'High'],
    'Calories': [450, 400, 520, 480, 300, 430]
}
df = pd.DataFrame(data)
df


## Filter Dishes Based on Calories

df[df['Calories'] > 400].sort_values(by = 'Calories')

## Filter Only North Indian Dishes

df[df['Type'] == 'North']

## Filter High Spice Dishes with Calories Over 400

df[(df['Spice_Level'] == 'High') & (df['Calories'] > 400)]


## Filter Dishes That Are Not Medium Spice

df[(df['Spice_Level'] != 'Medium')]

## Add a new column to your DataFrame


df['Meal_Time'] = ['Lunch', 'Breakfast', 'Dinner', 'Lunch', 'Dinner', 'Snack']
df

## Sort All Dinner Dishes by Calories

df[df['Meal_Time'] == 'Dinner'].sort_values(by = 'Calories')

```
